```
#df=pd.read_csv('tomslee_airbnb_new_york_0011_2014-05-10.csv')

final_df = pd.DataFrame()
import glob
path = "/content/drive/My Drive/MyWork/Airbnb New York/tomslee_airbnb_new_york_*.csv"
for filename in glob.glob(path):
    df = pd.DataFrame()
    df = pd.read_csv(filename)
    temp_df=df[['room_id', 'latitude', 'longitude', 'last_modified']]
    final_df=final_df.append(temp_df)
    print (filename, df.shape, final_df.shape, temp_df.shape)
```

```
import geopandas as gpd
```

```
gdf_map = gpd.read_file(gpd.datasets.get_path('nybb'))
```

```
gdf_map.set_index('BoroCode', inplace=True)
```

```
final_df['last_modified_fmtd'] = pd.to_datetime(final_df['last_modified']).dt.strftime('%Y%m')
final_df['last_modified_fmtd_Yr'] = pd.to_datetime(final_df['last_modified']).dt.strftime('%Y')
```

```
final_df_2014=final_df[final_df['last_modified_fmtd'].isin(['201412'])]
final_df_2015=final_df[final_df['last_modified_fmtd'].isin(['201512'])]
final_df_2016=final_df[final_df['last_modified_fmtd'].isin(['201612'])]
final_df_2017=final_df[final_df['last_modified_fmtd'].isin(['201707'])]
```

```
print (final_df_2014.shape, final_df_2015.shape, final_df_2016.shape, final_df_2017.shape)
```

```
gdf_data_2014 = gpd.GeoDataFrame(final_df_2014, geometry=gpd.points_from_xy(final_df_2014.longitude, final_df_2014.latitude))
gdf_data_2015 = gpd.GeoDataFrame(final_df_2015, geometry=gpd.points_from_xy(final_df_2015.longitude, final_df_2015.latitude))
gdf_data_2016 = gpd.GeoDataFrame(final_df_2016, geometry=gpd.points_from_xy(final_df_2016.longitude, final_df_2016.latitude))
gdf_data_2017 = gpd.GeoDataFrame(final_df_2017, geometry=gpd.points_from_xy(final_df_2017.longitude, final_df_2017.latitude))
```

```
print (gdf_data_2014.shape, gdf_data_2015.shape, gdf_data_2016.shape, gdf_data_2017.shape)
```
